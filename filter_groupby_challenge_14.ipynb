Sulfates cannot be higher than 0.6.
The price has to be less than $20.
Use the above conditions to filter the data for questions 2 and 3 below.

Questions:

```python
(df['region'] == 'Stellenbosch').value_counts()
```

```python
filtered_df = df[(df['sulphates'] <= 0.6) & (df['price'] < 20)]
```

```python
filtered_df.head()
```

```python
filtered_df.groupby(by = 'region')['price'].mean()
```

```python
print("Q2 avg price of wines from Bordeaux: ",filtered_df[filtered_df['region']=='Bordeaux']['price'].mean())

```
